```python
df = pd.read_csv("sat_sim_spatial.csv")
subset = (
    ((df["rowtype"] == "x") | (df["rowtype"] == "y") | (df["rowtype"] == "z"))
    & (df["vi_angle"] == 0.0)
    & (df["Apophis_mass"] == 2.2127565414999998e-20)
)
df_orbit = df[subset]
df_orbit = df_orbit.drop(columns=["vi_angle", "Apophis_mass"])
df_orbit = df_orbit.set_index("rowtype").T
df_orbit_Earth = df_orbit.iloc[:, ::4]
df_orbit_Earth = pd.concat([df_orbit_Earth.iloc[:, 0:3].drop("dt"), df_orbit_Earth.iloc[:, 3:].drop("dt")]).reset_index().drop(columns="index")
df_orbit_Ast = df_orbit.iloc[:, 1::4]
df_orbit_Ast = pd.concat([df_orbit_Ast.iloc[:, 0:3].drop("dt"), df_orbit_Ast.iloc[:, 3:].drop("dt")]).reset_index().drop(columns="index")
df_orbit_Sat1 = df_orbit.iloc[:, 2::4]
df_orbit_Sat1 = pd.concat([df_orbit_Sat1.iloc[:, 0:3].drop("dt"), df_orbit_Sat1.iloc[:, 3:].drop("dt")]).reset_index().drop(columns="index")
df_orbit_Sat2 = df_orbit.iloc[:, 3::4]
df_orbit_Sat2 = pd.concat([df_orbit_Sat2.iloc[:, 0:3].drop("dt"), df_orbit_Sat2.iloc[:, 3:].drop("dt")]).reset_index().drop(columns="index")
```
